```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name": ["Rahul", "Priya", "Amit", "Rahul", None],
    "Age": [22, np.nan, 30, 22, 28],
    "City": ["Mumbai", "Pune", None, "Mumbai", "Delhi"],
    "Salary": [35000, 50000, np.nan, 35000, 45000]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)
```

```python
print("\nAvailable Values")
print(df.notnull())
```

```python
print("\nFill Missing Age with Mean")
df["Age"] = df["Age"].fillna(df["Age"].mean())

print(df)

print("\nFill Missing Salary with 0")
df["Salary"] = df["Salary"].fillna(0)

print(df)

print("\nFill Missing City")
df["City"] = df["City"].fillna("Unknown")

print(df)
```

```python
print("\nDrop Rows with Missing Values")

print(df.dropna())
```

```python
print("\nDuplicate Rows")
print(df.duplicated())

print("\nDuplicate Count")
print(df.duplicated().sum())
```

```python
print("\nRemove Duplicates")
print(df.drop_duplicates())
```

```python
print("\nRename Columns")

df = df.rename(columns={
    "Age":"Employee_Age",
    "City":"Location"
})

print(df)
```

```python
print("\nChange Data Type")

df["Employee_Age"] = df["Employee_Age"].astype(float)

print(df.dtypes)
```

```python
print("\nReplace Values")

df["Location"] = df["Location"].replace({
    "Mumbai":"Pune",
    "Delhi":"Noida"
})

print(df)
```
