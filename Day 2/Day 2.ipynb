```python
import pandas as pd
```

```python
data = {
    "Name": ["Rahul", "Priya", "Amit", "Sneha", "Karan"],
    "Age": [22, 28, 30, 25, 35],
    "City": ["Mumbai", "Pune", "Delhi", "Mumbai", "Bangalore"],
    "Salary": [35000, 50000, 70000, 45000, 90000]
}

df = pd.DataFrame(data)

print(df)
```

```python
print(df["Name"])
```

```python
print(df[["Name", "Salary"]])
```

```python
print(df.head(3))
```

```python
print(df.loc[0])
```

```python
print(df.loc[1:3])
```

```python

print(df.loc[:, ["Name", "Salary"]])
```

```python
print(df.iloc[0:3])
```

```python
print(df[df["Age"] > 25])
```

```python
print(df[(df["Age"] > 25) & (df["Salary"] > 50000)])
```

```python
print(df[(df["City"] == "Mumbai") | (df["City"] == "Delhi")])
```

```python
print(df[~(df["City"] == "Mumbai")])
```

```python
print(df[~(df["City"] == "Mumbai")])
```

```python
print(df[df["Age"].between(25, 30)])
```

```python
print(df.query("Age > 25"))
```

```python
print(df.query("Age > 25 and Salary > 50000"))
```
