```python
import pandas as pd

print(pd.__version__)
```

```python
# Creating A index

data=["java","c++","ruby","go","jsp","julia","Python"]
# print(data)
    
for i in range(len(data)):
    print(i ,"=",data[i])
```

```python
# Series() Method

s=pd.Series(data)
print(s)
```

```python
data = ["Java", "C++", "Python", "Ruby", "Go", "jsp","julia"]
indexx = ["a", "b", "c", "d", "e", "f","g"]

s = pd.Series(data, index=indexx)
print(s)
```

```python
data={
    "a":100,"b":300,"c":400,"d":450
}

# print(data)

s=pd.Series(data)
print(s)
```
